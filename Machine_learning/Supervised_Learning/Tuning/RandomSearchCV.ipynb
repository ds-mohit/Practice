## Randomised Search Cv


````python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
````

````python
df = sns.load_dataset('Iris')
````

````python
df.head()
````

````python
df.info()
````

````python
df['species'].value_counts()
````

````python
from sklearn.model_selection import train_test_split
````

````python
X = df.drop('species', axis=1)
y = df['species']
````

````python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=101)
````

````python
from sklearn.neighbors import KNeighborsClassifier
````

````python
model_knn1 = KNeighborsClassifier(n_neighbors=13)
````

````python
model_knn1.fit(X_train, y_train)
````

````python
model_knn1.score(X_test, y_test)
````

````python
model_knn2 = KNeighborsClassifier(n_neighbors=5)
model_knn2.fit(X_train, y_train)
````

````python
model_knn2.score(X_test, y_test)
````

### here we can say that by changing the parameter we can get different results

````python
from sklearn.svm import SVC
````

````python
model_svm = SVC(C = 30, kernel='rbf',gamma = 'auto')
model_svm.fit(X_train, y_train)
````

````python
model_svm.score(X_test, y_test)
````

````python
model_svm1 = SVC(C = 30, kernel='linear',gamma = 'auto')
model_svm1.fit(X_train, y_train)
````

````python
model_svm1.score(X_test, y_test)
````

### here is the same thing that by changing the parameter we can get different results

### Now to find best Hyper Parameters we gona use ```Randomised Search CV```

````python
model_svm3 = SVC(gamma = 'auto')
model_svm3.fit(X_train, y_train)
````

````python
model_svm3.score(X_test, y_test)
````

````python
from sklearn.model_selection import RandomizedSearchCV
````

````python
classifier = RandomizedSearchCV((model_svm3),
                          {'C' : [1,10,20,30],
                           'kernel' : ['linear', 'rbf'],},
                           cv = 5,
                           n_iter=4,
                           return_train_score = False)
````

````python
classifier.fit(X,y) ## in this we directly give X and y data
````

````python
classifier.cv_results_
````

````python
results = pd.DataFrame(classifier.cv_results_)
results
````

````python
results[['param_C','param_kernel','mean_test_score']]
````
