### ``` Chi Square test ``` : 

the chi -square test is used to check if there is a significant association between two categorical variables. It helps to answer like :
1. is there a relationship between gender and voting preference ?
2. Do two different marketing strategies result in different customer prefrences ?
3. is there a difference between observed and expected frequencies in a survey or experiment ?

### ``` Chi Square test Formula ``` :

$$ x^2 = \frac {\sum (O_i-E_i)^2}{E_i}$$

$$
O_i = \text{observed frequency (actual data)}
$$

$$
E_i = \text{expected frequency (what we expect under the null hypothesis)}
$$

$$
\sum = \text{Sum of all categories}
$$


### ``` Steps ```:
1. null Hypothesis : there is no association ( or no difference ) between two categorical variables
2. Alternative Hypothesis : there is an association ( or difference ) between two categorical variables
3. Calculate Expected Frequencies (E) : for each cell in your data , the expected frequencies is calculated using :

$$
E_i = \frac{Row Total *Column Total}{Grand Total}
$$

4. calculate df (degree of freedom) : 
df = (r-1) * (c-1)

where r = number of rows and c = number of columns

by applying chi square test , we get chi values , then we need to find critical values using chi table for significance value(0.05) .

if chi value is less than critical value then we reject the null hypothesis .

if chi value is greater than critical value then we accept the null hypothesis .

````python
## chi-test using p-values:
import numpy as np
from scipy.stats import chi2_contingency
import seaborn as sns
import pandas as pd
````

````python
df = sns.load_dataset("titanic")
df
````

````python
## now we need to create a contigency table

ct = pd.crosstab(df["sex"] , df["survived"])
ct
````

````python
chi2 , p_value , dof , excepted = chi2_contingency(ct)

print("dof = " , dof)
print("p_value = " , p_value)
print("chi2 = " , chi2)
print("excepted = " , excepted)
````

````python
alpha = 0.05
if p_value < alpha:
    print("reject null hypothesis and there is a significance relationship between gender and survial")
else:
    print("accept null hypothesis and there is no relationship between gender and survial")
````
